```python
import openai
from openai import OpenAI
import os
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import gradio as gr
import ast
```

```python
api_key_path = "Documents/Vinika_Personal/MYOPENAI_Api_key.txt"
```

```python
def read_api_key(file_path):
    with open(file_path, 'r') as file:
        text = file.read().strip()
#         print(text)
        return text
```

```python
os.environ["OPENAI_API_KEY"] = read_api_key(api_key_path)
```

```python
openai.api_key = os.environ["OPENAI_API_KEY"]  # Replace with your OpenAI API key
EMBEDDING_MODEL = "text-embedding-ada-002"
LLM_MODEL = "gpt-3.5-turbo"
```

```python
def load_data():
    # Replace this with your real dataset
    data = {
        "Product Name": ["HydraBoost Gel", "ClearSkin Serum", "Soothing Night Cream"],
        "Brand": ["GlowSkin", "DermaFix", "SilkDerm"],
        "Description": [
            "Lightweight hydrating gel with hyaluronic acid for dry skin.",
            "BHA + niacinamide serum to reduce acne and control oil.",
            "Rich night cream with ceramides and aloe for sensitive skin."
        ]
    }
    df = pd.DataFrame(data)
    return df
```

```python
df = load_data()
```

```python
client = OpenAI(api_key=openai.api_key)
```

```python
def get_embedding(text):
    response = client.embeddings.create(
        input=[text],
        model=EMBEDDING_MODEL
    )
    return response.data[0].embedding

# Generate or load embeddings
def generate_embeddings(df):
    if 'embedding' not in df.columns:
        df['embedding'] = df['Description'].apply(get_embedding)
    return df
```

```python
# pip install --upgrade openai ----ALREADY Upgraded
```

```python
df = generate_embeddings(df)
```

```python
def search_products(user_query, df):
    user_embedding = get_embedding(user_query)
    df['score'] = df['embedding'].apply(lambda x: cosine_similarity([user_embedding], [x])[0][0])
    return df.sort_values('score', ascending=False).head(3)
```

```python
def explain_recommendations(user_query, top_products):
    products_text = "\n".join(
        [f"{row['Product Name']} by {row['Brand']} - {row['Description']}" for _, row in top_products.iterrows()]
    )

    prompt = f"""
    A user is looking for skincare products for: "{user_query}".
    Based on the following options, recommend the top 3 with reasons:

    {products_text}

    Return the recommendations as bullet points.
    """

    response = openai.ChatCompletion.create(
        model=LLM_MODEL,
        messages=[{"role": "user", "content": prompt}]
    )

    return response['choices'][0]['message']['content']
```

```python
def recommend(user_query):
    top = search_products(user_query, df)
    return explain_recommendations(user_query, top)
```

```python
gr.Interface(fn=recommend, inputs="text", outputs="text", title="Skincare Product Recommender").launch()
```
